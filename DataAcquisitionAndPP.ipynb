```python
outFrame =frame[frame.tic.isin(['ADA-USD' ,  'BNB-USD',   'BTC-USD',   'ETH-USD' ,  'LTC-USD'  , 'XLM-USD'])]
```

```python
outFrame.head()
```

```python
#Getting covariance matrices

outFrame = outFrame.sort_values(['date','tic'])
outFrame.index = outFrame.date.factorize()[0]

covarianceList = []
listOfReturns = []


#Lookback is set to one year. 365??
lookbackRange = 365

for i in range (lookbackRange, len(outFrame.index.unique())):
  frame_lookBack = outFrame.loc[i-lookbackRange:i,:]
  pivotFrame = frame_lookBack.pivot_table(index='date', columns= 'tic',values= 'close')
  lookbackReturns = pivotFrame.pct_change().dropna()
  listOfReturns.append(lookbackReturns)

  covarObj = lookbackReturns.astype(float).cov().values
  print(covarObj)
  covarianceList.append(covarObj)




covFrame = pd.DataFrame({'date':outFrame.date.unique()[lookbackRange:], 'covarianceList': covarianceList, 'listOfReturns':listOfReturns}) 



```

```python

outFrame= outFrame.merge(covFrame, on='date')
outFrame = outFrame.sort_values(['date','tic']).reset_index(drop = True)
outFrame.head()
```
